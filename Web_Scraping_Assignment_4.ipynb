QUESTION 1

```python
'''Scrape the details of most viewed videos on YouTube from Wikipedia. Url 
= https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details: 
A) Rank 
B) Name 
C) Artist 
D) Upload date 
E) Views'''
import selenium
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException,StaleElementReferenceException
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import requests
import warnings
warnings.filterwarnings("ignore")
import time
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
```

```python
ranks1 = []
ranks = driver.find_elements(By.XPATH,'//table[@class = "wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[1]')
for i in ranks:
    rank = i.text
    ranks1.append(rank)
```

```python
names1 = []
names = driver.find_elements(By.XPATH,'//table[@class = "wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[2]/a')
for i in names:
    name = i.text
    names1.append(name)
names1.insert(11,'Masha and the Bear – Recipe for Disaster')
```

```python
artists1 = []
artists = driver.find_elements(By.XPATH,'//table[@class = "wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[3]')
for i in artists:
    artist = i.text
    artists1.append(artist)
```

```python
views1 = []
views = driver.find_elements(By.XPATH,'//table[@class = "wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[4]')
for i in views:
    view = i.text
    views1.append(view)
```

```python
upload_date1 = []
upload_date = driver.find_elements(By.XPATH,'//table[@class = "wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[5]')
for i in upload_date:
    u_p = i.text
    upload_date1.append(u_p)
```

```python
import pandas as pd
df = pd.DataFrame({'Ranks':ranks1,'Name of the Video':names1,'Artists':artists1,'Views(in billions)':views1,'Upload Date':upload_date1})
df
```

```python
dataset_name = []
start = 0
end = 26
try:
    for page in range(start,end):
        d_name = driver.find_elements(By.XPATH,'//h2[@class = "truncate text-primary"]/a')
        for i in d_name:
            d_n = i.text
            dataset_name.append(d_n)
        next_button = driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[3]/div/button[2]')
        next_button.click()
        time.sleep(7)
except NoSuchElementException:
    dataset_name.append('-')   
```

```python
data_type = []
start = 0
end = 26
try:
    for page in range(start,end):
        d_type = driver.find_elements(By.XPATH,'//div[@class = "my-2 hidden gap-4 md:grid grid-cols-12"]/div[2]/span')
        for i in d_type:
            d_t = i.text
            data_type.append(d_t)
        next_button = driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[3]/div/button[2]')
        next_button.click()
        time.sleep(7)
except NoSuchElementException:
    data_type.append('-')   
```
